```python
import pandas as pd
import numpy as np
from scipy import stats
from google.colab import files
import io

# Solicitar al usuario que suba un archivo de Excel
def cargar_datos_desde_excel():
    print("Por favor, selecciona un archivo Excel con los datos de tu encuesta.")
    archivo_subido = files.upload()  # Cargar archivo
    try:
        for archivo_nombre in archivo_subido.keys():
            df = pd.read_excel(io.BytesIO(archivo_subido[archivo_nombre]))
            return df
    except Exception as e:
        print(f"Error al cargar el archivo: {e}")
        return cargar_datos_desde_excel()

dataframe = cargar_datos_desde_excel()

# Seleccionar columnas específicas para análisis
preguntas = [
    "1. ¿Cómo calificarías la gestión de la CBN durante la pandemia del COVID-19?",
    "2. ¿ ¿Crees que las medidas implementadas por la CBN durante la pandemia te ayudaron a sentirte más seguro en tu lugar de trabajo?",
    "4. ¿Te sientes satisfecho con las políticas laborales implementadas por la CBN durante los momentos de crisis?"
]

resultados = {}

for pregunta in preguntas:
    if pregunta in dataframe.columns:
        datos = dataframe[pregunta].dropna()
        # Convertir datos a categóricos si no son numéricos
        if datos.dtypes == 'object':
            datos = datos.astype('category').cat.codes

        media = np.mean(datos)
        mediana = np.median(datos)
        moda = stats.mode(datos, keepdims=True)[0][0]

        resultados[pregunta] = {
            "Media": media,
            "Mediana": mediana,
            "Moda": moda
        }
    else:
        print(f"La pregunta '{pregunta}' no se encontró en el archivo.")

# Mostrar resultados
print("\nResultados de las Medidas de Tendencia Central:")
for pregunta, medidas in resultados.items():
    print(f"\n{pregunta}")
    print(f"Media: {medidas['Media']}")
    print(f"Mediana: {medidas['Mediana']}")
    print(f"Moda: {medidas['Moda']}")

# Visualización básica (opcional)
import matplotlib.pyplot as plt

for pregunta in preguntas:
    if pregunta in dataframe.columns:
        datos = dataframe[pregunta].dropna()
        if datos.dtypes == 'object':
            datos = datos.astype('category').cat.codes

        plt.figure()
        plt.hist(datos, bins=10, color='skyblue', edgecolor='black')
        plt.title(f"Distribución de Respuestas: {pregunta}")
        plt.xlabel("Valores")
        plt.ylabel("Frecuencia")
        plt.show()
```
